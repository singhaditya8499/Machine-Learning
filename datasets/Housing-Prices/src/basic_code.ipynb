## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
data = pd.read_csv("../data/data.csv")
```

```python
data.head(3)
```

## Column description

1. **CRIM** - per capita crime rate by town
2. **ZN** - proportion of residential land zoned for lots over 25,000 sq.ft.
3. **INDUS** - proportion of non-retail business acres per town
4. **CHAS** - Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
5. **NOX** - nitric oxides concentration (parts per 10 million)
6. **RM** - average number of rooms per dwelling
7. **AGE** - proportion of owner-occupied units built prior to 1940
8. **DIS** - weighted distances to five Boston employment centres
9. **RAD** - index of accessibility to radial highways
10. **TAX** - full-value property-tax rate per \$10,000

11. **PTRATIO** - pupil-teacher ratio by town
12. **B** - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks
by town
13. **LSTAT** -  % lower status of the population
14. **MEDV** - Median value of owner-occupied homes in $1000's

```python
data.isna().sum()
```

This shows that there are 5 NA or NULL values in RM column of the dataset.

```python
data.describe().T
```

This confirms our fininding further. We can see that RM has 5 null values, we will fill that value with the mean of the columns RM.

```python
mean_value = data['RM'].mean()
 
data['RM'].fillna(value=mean_value, inplace=True)
```

```python
data.describe().T
```

```python
# normalising the dataset with min-max normalising
scaled_data = (data - data.min())/(data.max() - data.min())
```

```python
scaled_data
```

```python
# splitting the data into test and training set
Y = scaled_data.MEDV
X = scaled_data.drop(columns=['MEDV'])
```

```python
print("X shape: ", X.shape)
print("Y shape: ", Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=37)
```

## Neural network

```python
import tensorflow as tf
from tensorflow.keras import layers, optimizers
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(layers.Dense(128, activation='relu', input_shape=(13, )))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(1))

optimizer = optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=500, validation_split=0.2, verbose=False)
```

```python
plt.figure(figsize = (10,8))
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Epochs vs Loss")
plt.grid()
plt.legend()
plt.show()
```

Looking at the `Loss vs Epochs` curve we can go with epoch=100 for this problem.

```python
model_final = Sequential()
model_final.add(layers.Dense(128, activation='relu', input_shape=(13, )))
model_final.add(layers.Dense(64, activation='relu'))
model_final.add(layers.Dense(32, activation='relu'))
model_final.add(layers.Dense(4, activation='relu'))
model_final.add(layers.Dense(1))

optimizer = optimizers.Adam(learning_rate=0.001)
model_final.compile(optimizer=optimizer, loss='mean_squared_error')
```

```python
model_final.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
model_final.evaluate(X_test, y_test)
```

```python
prediction = model_final.predict(X_test)
result = pd.DataFrame({'actual': y_test, 'predicted': np.reshape(prediction, (103,))})
result['MSE'] = ((result.actual - result.predicted)**2)/float(result.shape[0])
result
```
