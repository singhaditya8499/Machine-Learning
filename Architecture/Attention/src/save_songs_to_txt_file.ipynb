```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/cleaned-data-with-features.csv")
```

```python
data.head(3)
```

```python
songs = data['song_name'].values
lyrics = data['lyrics'].values
combined = [f"Song Name: {song}\n{lyric}\n" for song, lyric in zip(songs, lyrics)]

all_songs = "\n-----\n".join(combined)
with open("../data/songs.txt", "w") as f:
    f.write(all_songs)
```
